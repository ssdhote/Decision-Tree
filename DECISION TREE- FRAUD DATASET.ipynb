```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets  
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
data=pd.read_csv("C:\\Users\\DELL\\Downloads\\Fraud_check.csv")
data.head()
```

```python
data.describe()
```

```python
label_encoder = preprocessing.LabelEncoder()
data['Undergrad']= label_encoder.fit_transform(data['Undergrad']) 
data['Marital.Status']= label_encoder.fit_transform(data['Marital.Status']) 
data['Urban']= label_encoder.fit_transform(data['Urban'])
```

```python
data.rename({'Taxable.Income':"Taxable_income"},axis=1,inplace=True)
data.head()
```

```python
bins=[0,30000,100000]
name=["risky","Not_risky"]
data["Tax"]=pd.cut(data.Taxable_income,bins,labels=name)
data.head()
```

```python
data.drop("Taxable_income",axis=1,inplace=True)
data['Tax']= label_encoder.fit_transform(data['Tax'].astype(str))
data.head()
```

```python
x=data.iloc[:,0:5]
y=data.iloc[:,5]
```

```python
x
```

```python
y
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=40)
```

```python
model = DecisionTreeClassifier(criterion = 'entropy')
model.fit(x_train,y_train)
```

```python
tree.plot_tree(model);
```

```python
fn=['Undergrad','Marital.Status','City.Population','Work.Experience','Urban']
cn=['risky', 'Not_risky']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(model,
               feature_names = fn, 
               class_names=cn,
               filled = True);
```

```python
preds = model.predict(x_test) # predicting on test data set 
pd.Series(preds).value_counts() # getting the count of each category 
```

```python
pd.crosstab(y_test,preds)
```
